- 지도 마커표시
    - Marker() 클래스로 생성한 뒤 지도에 추가
    - Icon() 내 들어가는 icon 속성값은 https://getbootstrap.com/docs/3.3/components/ 에서 바꾼다

```python
## 마커
map = folium.Map(location=[35.178508, 129.199443], tiles='CartoDB Positron', zoom_start=16)

folium.Marker(location=[35.178508, 129.199443], tooltip='송정 해수욕장', popup='서핑하러 오세요~!!').add_to(map)
popup=folium.Popup('<pre><a href="https://www.starbucks.co.kr/">스타벅스로 오세요</a></pre>',min_width=200, max_width=300)
folium.Marker(location=[35.180762, 129.202324], tooltip='스타벅스 부산송정비치점', popup=popup, icon=folium.Icon(color='green',icon='pushpin')).add_to(map)
map
```
